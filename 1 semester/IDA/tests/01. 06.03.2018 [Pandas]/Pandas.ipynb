```python
import pandas as pd
import numpy as np
import datetime

from IPython.display import display
```

```python
regions = pd.read_csv('salaries-by-region.csv')
college_type = pd.read_csv('salaries-by-college-type.csv', index_col=0)
```

```python
regions['SMS'] = regions['Starting Median Salary'].str.strip('$').str.replace(',', '').astype(float)
regions.head()
```

```python
college_type['SMS'] = college_type['Starting Median Salary'].str.strip('$').str.replace(',', '').astype(float)
college_type.head()
```

```python
regions.Region.value_counts() #1
```

```python
college_type['School Type'].value_counts()['State'] #2
```

```python
college_type[college_type['School Type'] == 'Engineering'].SMS.quantile(q=0.7)
```

```python
college_type[college_type.SMS == college_type.SMS.max()]
```

```python
display(college_type)
```

```python
college_type.groupby('School Type').agg({'SMS': 'mean'}).round(0)
```

```python
regions['MC90PS'] = regions['Mid-Career 90th Percentile Salary'].str.strip('$').str.replace(',', '').astype(float)
display(regions)
```

```python
regions.groupby('Region').agg({'MC90PS': ['median', 'std', 'min', 'max']}).round(0)
```

```python
college_type = college_type.reset_index()
```

```python
college_region = college_type.merge(regions[['School Name', 'Region']], on='School Name')
```

```python
college_region
```

```python
college_region['MC25PS'] = college_region['Mid-Career 25th Percentile Salary'].str.strip('$').str.replace(',', '').astype(float)
```

```python
college_region.MC25PS
```

```python
college_region[college_region.Region == 'California'].groupby('School Type').agg({'MC25PS': 'mean'}).round(0)
```
