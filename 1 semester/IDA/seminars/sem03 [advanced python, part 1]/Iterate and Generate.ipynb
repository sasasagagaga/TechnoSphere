### Как конвертировать циклы в генераторы?

```python
numbers = [1, 2, 3, 4, 5, 6]

odds_2 = []
for n in numbers:
    if n % 2 == 1:
        odds_2.append(2 * n)

odds_2
```

```python
numbers = [1, 2, 3, 4, 5, 6]        

odds_2 = [2 * n for n in numbers if n % 2 == 1]

odds_2
```

## Функции-генераторы

```python
def triangle(n):
    n = n + 1
    for i in range(1, n):
        yield ''.join('*' if n - i < j < n + i else ' ' for j in range(2 * n))

print(*triangle(5), sep='\n', end='')
```

```python
def gen_func():
    for w in 'abc':
        for f in '123':
            yield w + '-' + f
            
gen_func()
```

```python
gen = gen_func()

print(next(gen))
print(next(gen))
print(next(gen))
```

```python
for i in gen_func():
    print(i)
```

```python
def accumulator():
    total = 0
    while True:
        value = yield total
        print(f"Accepted: {value}")
        
        if not value:
            break
        else:
            total += value
    yield total
```

```python
gen = accumulator()

print('Sum: {}'.format(next(gen)))
print('Sum: {}'.format(gen.send(1)))
print('Sum: {}'.format(gen.send(2)))

next(gen)
```

```python
def gen_squares():
    return (i ** 2 for i in range(3))

def complex_gen():    
    for i in "sphere":
        yield 'Letter:', i
        yield from gen_squares()
```

```python
for i in complex_gen():
    print(i)
```
